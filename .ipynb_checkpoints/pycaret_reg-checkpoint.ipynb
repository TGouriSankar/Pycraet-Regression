```python
import numpy as np
import pandas as pd
import re
import warnings
warnings.filterwarnings("ignore")
```

```python
bike = pd.read_csv("Avenger_Assemble.csv")
bike
```

```python
bike.drop(columns=['Unnamed: 0'],axis=1,inplace=True)
```

```python
bike
```

```python
bike[bike.duplicated()]
```

```python
bike = bike.drop_duplicates().reset_index(drop=True)
```

```python
for i in bike.columns:
    print(i,":",bike[i].value_counts())
    print("- "*40)
```

```python
for i in bike.columns:
    print(i,":",bike[i].unique())
    print("- "*40)
    print("- "*40)
```

```python
for i in bike.columns:
    print(i,":",bike[i].nunique(),"-", bike[i].dtypes)
    print("- "*50)
```

```python
total_null = bike.isnull().sum()
total_null
```

```python
total_num = np.product(bike.shape)
total_miss = total_null.sum()
percentage = (total_miss/total_num)*100
print("The percentage of missing values",percentage)
```

```python
def cleaner(num):
    try:
        m = re.sub('[^0-9]',"",num)
        return m
    except:
        pass
```

```python
bike['bike_price'] = bike['bike_price'].apply(lambda x : cleaner(x))
bike['km_driven'] = bike['km_driven'].apply(lambda x : cleaner(x))
```

```python
bike['bike_price'] = bike['bike_price'].astype('float64')
bike['km_driven'] = bike['km_driven'].astype('float64')
```

```python
bike.head()
```

```python
bike.dtypes
```

```python
bike['bike_address'] = bike['bike_address'].str.split(",").str.get(0)
```

```python
bike['bike_name'] = bike['bike_name'].str.lower()
bike['bike_address'] = bike['bike_address'].str.lower()
bike['state'] = bike['state'].str.lower()
```

```python
bike
```

```python
num_data = bike.select_dtypes(include=np.number)
cat_data = bike.select_dtypes(include="object")
```

```python
num_data
```

```python
cat_data
```

```python
for i in num_data.columns:
    x = [num_data[i].isnull()]
    if x==0:
        pass
    else:
        num_data[i].fillna(num_data[i].median(),inplace=True)
```

```python
for i in cat_data.columns:
    x = [cat_data[i].isnull()]
    if x==0:
        pass
    else:
        cat_data[i].fillna(cat_data[i].mode()[0],inplace=True)
```

```python
bike = pd.concat([num_data,cat_data],axis=1)
bike
```
